1. ¿Qué tan a menudo se retrasan los vuelos en comparación a la cantidad de estos por aerolína?

```python
%%sql 

SELECT o.Reporting_Airline, (((SELECT COUNT(*) FROM ontime o1 WHERE o1.DepDelayMinutes!=0 and o1.Month>=9 and o1.Year=2018 AND o.Reporting_Airline = o1.Reporting_Airline)/3.0 )/


		((SELECT COUNT(*) FROM ontime o4 WHERE o4.Month>=9 and o4.Year=2018 AND o.Reporting_Airline = o4.Reporting_Airline) /3.0))*100 AS "FRECUENCIA"

FROM ontime o
GROUP BY o.Reporting_Airline
ORDER BY FRECUENCIA ASC
```

2. ¿Qué proporción de vuelos se cancelan debido al código A (Carrier)?


```python
%%sql
SELECT c.Description as AEROLINEA,CAST (count (*) AS float) / ( 
  SELECT count(*) 
  FROM ontime o1 
  WHERE o1.Cancelled = 0 
    AND o1.Reporting_Airline = o.Reporting_Airline 
  ) * 100 AS PROPORCION 
FROM ontime o 
INNER JOIN carrier_history c ON o.Reporting_Airline = c.Code 
WHERE Cancelled = 1 
  AND CancellationCode = 'A' 
GROUP BY o.Reporting_Airline 
ORDER BY PROPORCION ASC;
```

3. ¿Cuál es el promedio de vuelos mensuales que realizan las aerolíneas?

```python
%%sql
SELECT o.Reporting_Airline, ((SELECT COUNT(*) FROM ontime o1 WHERE o1.Year=2018 AND o.Reporting_Airline = o1.Reporting_Airline)/11)  as "PROMEDIO"
FROM ontime o
GROUP BY o.Reporting_Airline
ORDER BY PROMEDIO DESC
```

4. ¿Qué aerolíneas tienen la mayor cantidad de destinos?

```python
%%sql
SELECT c.Description, count(DISTINCT o.DestCityName) as TOTAL 
FROM ontime o 
INNER JOIN carrier_history c ON o.Reporting_Airline = c.Code 
GROUP BY c.Code 
ORDER BY TOTAL DESC;
```


5. ¿Qué tan puntuales son los vuelos de las aerolíneas en comparación a la cantidad de viajes?

```python
%%sql
SELECT o.Reporting_Airline, (((SELECT SUM(o1.DepDelayMinutes) FROM ontime o1 WHERE o.Reporting_Airline = o1.Reporting_Airline and o1.Year=2018 GROUP BY o1.Month,o1.DayofMonth) +
       (SELECT SUM(o2.ArrDelayMinutes) FROM ontime o2 WHERE o.Reporting_Airline = o2.Reporting_Airline and o2.Year=2018 GROUP BY o2.Month,o2.DayofMonth))/334) AS "MINUTOS_DE_RETRASO_POR_DIA",
		((SELECT COUNT() FROM ontime o3 WHERE o.Reporting_Airline = o3.Reporting_Airline and o3.Year=2018 GROUP BY o3.Month,o3.DayofMonth)*1.0/334) AS "CANT_VUELOS_POR_DIA"
		
FROM ontime o
GROUP BY o.Reporting_Airline
ORDER BY MINUTOS_DE_RETRASO_POR_DIA ASC, CANT_VUELOS_POR_DIA DESC
```

6. ¿Qué tan frecuentemente un vuelo sale tarde por culpa propia de la aerolínea en comparación con la cantidad de vuelos?

```python
%%sql
SELECT c.Description as AEROLINEA,CAST (count (*) AS float) / ( 
  SELECT count(*) 
  FROM ontime o1 
  WHERE o1.Cancelled = 0 
    AND o1.Reporting_Airline = o.Reporting_Airline 
  ) * 100 AS PROPORCION 
FROM ontime o 
INNER JOIN carrier_history c ON o.Reporting_Airline = c.Code 
WHERE o.CarrierDelay > 0 AND o.DepDelay > 0 
  AND o.Cancelled = 0 
GROUP BY o.Reporting_Airline 
ORDER BY PROPORCION ASC;
```

7. ¿Qué aerolíneas han tenido un aumento en la realización de sus vuelos a lo largo del 2018?

```python
%%sql
SELECT o.Reporting_Airline, (((SELECT COUNT(*) FROM ontime o1 WHERE o1.Month=11 and o1.Year=2018 AND o.Reporting_Airline = o1.Reporting_Airline) - 
							(SELECT COUNT(*) FROM ontime o2 WHERE o2.Month=1 and o2.Year=2018 AND o.Reporting_Airline = o2.Reporting_Airline))*1.0 /
							(SELECT COUNT(*) FROM ontime o3 WHERE o.Reporting_Airline = o3.Reporting_Airline AND o3.Year=2018  GROUP BY o3.Month ORDER BY  COUNT(*) DESC LIMIT 1))*100 AS "PROPORCION_DE_AUMENTO"
FROM ontime o
GROUP BY o.Reporting_Airline
ORDER BY PROPORCION_DE_AUMENTO DESC
```

8. ¿Qué aerolíneas recorren distancias más grandes?

```python
%%sql
SELECT ch.Description as AEROLINEA, MAX(o.Distance) as DIST 
FROM carrier_history ch 
INNER JOIN ontime o ON ch.Code = o.Reporting_Airline 
GROUP BY ch.Code 
ORDER BY DIST DESC;
```

9. ¿Qué aerolíneas tienen más vuelos en un día durante el 2018?


```python
%%sql
SELECT o.Reporting_Airline, ((SELECT COUNT() FROM ontime o3 WHERE o.Reporting_Airline = o3.Reporting_Airline and o3.Year=2018 GROUP BY o3.Month,o3.DayofMonth)*1.0/334) AS "CANT_VUELOS_POR_DIA"
FROM ontime o
GROUP BY o.Reporting_Airline
ORDER BY CANT_VUELOS_POR_DIA DESC
```

10. ¿Qué tan activa ha sido una aerolínea en los últimos tres meses?


```python
%%sql
SELECT ch.Description as AEROLINEA, CAST (count(*)/3 as float) as PROM_3M 
FROM carrier_history ch 
INNER JOIN ontime o ON ch.Code = o.Reporting_Airline 
WHERE o.Year = 2018 AND o.Month >= 9 
GROUP BY o.Reporting_Airline 
ORDER BY PROM_3M DESC;
```
